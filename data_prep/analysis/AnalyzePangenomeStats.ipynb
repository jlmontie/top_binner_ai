```python
import plotly.express as px
import pandas as pd
```

```python
pan_stats = pd.read_csv('pangenome_expansion_stats.csv')
```

```python
fig = px.box(pan_stats, y='Pangenome expansion factor', hover_data=['Organism', 'Pangenome size', 'taxid reference genomes'])
fig.show()
```

```python
fig2 = px.scatter(pan_stats, x='taxid reference genomes', y='Pangenome expansion factor',
                  hover_data=['Organism', 'Pangenome size', 'taxid reference genomes'])
fig2.show()
```

```python
fig3 = px.box(pan_stats, y='Pangenome size', hover_data=['Organism', 'Pangenome size', 'taxid reference genomes'])
fig3.show()
```

```python
fig4 = px.scatter(pan_stats, x='Pangenome size', y='Pangenome expansion factor',
                  hover_data=['Organism', 'Pangenome size', 'taxid reference genomes'])
fig4.show()
```
